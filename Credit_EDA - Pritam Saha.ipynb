```python
import warnings

warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
df = pd.read_csv('application_data.csv')
```

```python
df.head()
```

```python
df.info(verbose = True) 
```

```python
df.shape
```

```python
df.describe()
```

```python
# Percentage of missing values in all columns

df.isnull().mean()*100
```

```python
# dropping columns with missing percentage greater than 50%

max_missing_cols = df.columns[df.isnull().mean()>0.5]
df.drop(max_missing_cols, axis=1, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

AMT_REQ_CREDIT_BUREAU_HOUR, AMT_REQ_CREDIT_BUREAU_DAY, AMT_REQ_CREDIT_BUREAU_WEEK, AMT_REQ_CREDIT_BUREAU_MON, AMT_REQ_CREDIT_BUREAU_QRT, AMT_REQ_CREDIT_BUREAU_YEAR are the columns whose missing percentage is around 13%. We can use the median value of the respective columns to impute the missing value.


```python
# Checking the data types of all the columns

df.dtypes
```

```python
# Changing the negative values to absolute values for ease of calculation in the below columns.

df.DAYS_BIRTH = abs(df.DAYS_BIRTH)
df.DAYS_REGISTRATION = abs(df.DAYS_REGISTRATION)
df.DAYS_EMPLOYED = abs(df.DAYS_EMPLOYED)
df.DAYS_ID_PUBLISH = abs(df.DAYS_ID_PUBLISH)
df.DAYS_LAST_PHONE_CHANGE = abs(df.DAYS_LAST_PHONE_CHANGE)
```

```python
df.head()
```

```python
# Analysing outlier of DAYS_EMPLOYED column

plt.figure(figsize = (10,5))
sns.boxplot(df.DAYS_EMPLOYED)
plt.show()

# The column has values like 365243, which is an outlier. It can never be possible because 365243/365 = 1000 years(appox).
```

```python
# Analysing outlier of the income column

plt.figure(figsize = (10,5))
sns.boxplot(df.AMT_INCOME_TOTAL)
plt.show()

# Some observations are very far from the normal observations, which are ouliers. It might be due to wrong entry or a customer with different profile
```

```python
# Analysing the outlier of the REGION_POPULATION_RELATIVE column

plt.figure(figsize = (10,5))
sns.boxplot(df.REGION_POPULATION_RELATIVE)
plt.show()

# There is outlier in the column which may be due to wrong entry.
```

```python
# Analysing outlier in Days_Registration column

plt.figure(figsize = (10,5))
sns.boxplot(df.DAYS_REGISTRATION)
plt.show()

# Lots of outliers are observed above the upper quartile range.
```

```python
# Analysing outlier of the  DAYS_LAST_PHONE_CHANGE column

plt.figure(figsize = (10,5))
sns.boxplot(df.DAYS_LAST_PHONE_CHANGE)
plt.show()
```

```python
# Creating a new AGE column from the days birth column by rounding off the result and typecasted it to integer.

df['AGE'] = round(df['DAYS_BIRTH']/365).astype(int)
```

```python
df.head()
```

```python
# Binning the Days_Birth column to find the applicants age group.

pd.cut(df['AGE'], bins = 3, labels = ['teenage', 'young', 'senior_citizen']).value_counts().plot(kind = 'bar')
plt.show()
```

```python
# Binning the external source 2 column to understand the rating from the source

pd.cut(df['EXT_SOURCE_2'], bins = 3, labels = ['low', 'mid', 'high']).value_counts().plot(kind = 'bar')
plt.show()
```

```python
# Binning the external source 3 column to understand the rating from the source

pd.cut(df['EXT_SOURCE_3'], bins = 3, labels = ['low', 'mid', 'high']).value_counts().plot(kind = 'bar')
plt.show()
```

```python
# Checking the imbalance percentage of the Target Column

df['TARGET'].value_counts()*100/len(df)

# It is found imbalanced
```

```python
# Dividing the application data set into two data sets on the basis of target column

df_1 = df[df['TARGET']==1]
df_0 = df[df['TARGET']==0]
```
